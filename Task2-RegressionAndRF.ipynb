```python
DATADIR='./dataset1/'

# Copy from previous notebook
def create_sub(preds, fn=None):
    df =  pd.DataFrame({'Id': range(len(preds)), 'Expected': preds})
    if fn is not None: df.to_csv(DATADIR + fn, index=False)
    return df
```

```python
def print_scores_st(m, station, X_train=X_train, X_valid=X_valid):
    yv = y_valid.loc[X_valid['station_alt'] == station]
    yt = y_train.loc[X_train['station_alt'] == station]
    print('Train R2 = ', m.score(X_train, yt), 
          ', Valid R2 = ', m.score(X_valid, yv),', Valid MSE = ', mse(m.predict(X_valid), yv))
```
